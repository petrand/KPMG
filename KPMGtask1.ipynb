```python
def age_group(age):
    ageGroup = ''
    if age<20:
        ageGroup = '0-20'
    elif age<30:
        ageGroup = '20-30'
    elif age<40:
        ageGroup = '30-40'
    elif age<50:
        ageGroup = '40-50'
    elif age<60:
        ageGroup = '50-60'
    elif age<80:
        ageGroup = '60-80'
    else:
        ageGroup = '80+'
    return ageGroup
```

```python
customer_demographic['ageGroup'] = customer_demographic.apply(lambda row: age_group(row['age']), axis=1)
customer_demographic.head()
```

```python
customer_demographic[['past_3_years_bike_related_purchases', 'ageGroup']].boxplot(by='ageGroup')
for ageGroup in customer_demographic.ageGroup.unique():
    mean = customer_demographic[customer_demographic['ageGroup'] == ageGroup]['purchases'].mean()
    print(mean)
```

```python
def total_purchase(transactions):
    total_cost_dict = {}
    for customer_id in transactions.customer_id.unique():
        purchases = transactions.loc[transactions['customer_id']==customer_id]
        total_cost = purchases['list_price'].sum()
        total_cost_dict[int(customer_id)] = int(total_cost)
    keys = sorted(total_cost_dict.keys())
    purchases = [int(total_cost_dict[i]) for i in keys]
    return pd.Series({'customer_id': keys, 'purchases': purchases})
```

```python
purchase_series = total_purchase(transactions)
len(purchase_series['customer_id'])
```

```python
dummy = [0 for i in range(0,customer_demographic.shape[0])]
customer_demographic['purchases'] = dummy
customer_demographic.head()
```

```python
def gender_fix(x):
    if x == 'F' or x == 'Femal':
        x = 'Female'
    if x=='M':
        x='Male'
    return x
def NSW_fix(x):
    if x=='New South Wales':
        x='NSW'
    if x=='Victoria' or x=='Vic':
        x='VIC'
    return x
customer_demographic['state'] = customer_demographic.apply(lambda x: NSW_fix(x['state']), axis=1)
customer_demographic['gender'] = customer_demographic.apply(lambda x: gender_fix(x['gender']), axis=1)
```

```python
import json
import time
import requests
def get_latlng(address, state, postcode, country):
    address = address + ", " + state + ", " + str(postcode) + ', ' + country
    key = {"key" : "40c1ae6866a363"}
    url = "https://us1.locationiq.com/v1/search.php"
    params = {
        'key' : key["key"],
        'q' : address,
        'format' : 'json'
    }
    time.sleep(0.2)
    response = requests.get(url, params=params)
    
    if response.status_code != 200:
            return pd.Series({'Latitude' : response.status_code,  'Longitude' : response.status_code})
    
    lat = response.json()[0]['lat']
    lng = response.json()[0]['lon']
    return pd.Series({'Latitude': lat, 'Longitude':lng})
```

```python
customer_demographic[['Latitude', 'Longitude']] = customer_demographic.apply(lambda x: get_latlng(x['address'], x['state'],x['postcode'], x['country']), axis=1)
```

```python
def total_std_cost(transactions):
    total_cost_dict = {}
    for customer_id in transactions.customer_id.unique():
        purchases = transactions.loc[transactions['customer_id']==customer_id]
        total_cost = purchases['standard_cost'].sum()
        total_cost_dict[int(customer_id)] = int(total_cost)
    keys = sorted(total_cost_dict.keys())
    purchases = [int(total_cost_dict[i]) for i in keys]
    return pd.Series({'customer_id': keys, 'std_cost': purchases})
std_cost = total_std_cost(transactions)
std_cost
```

```python
dummy = [0 for i in range(0,customer_demographic.shape[0])]
customer_demographic['std_cost'] = dummy
customer_demographic.head()
```

```python
for index, row in customer_demographic.iterrows():
    customer_id = row['customer_id']
    try:
        x = std_cost['std_cost'][customer_id-1]
        customer_demographic['std_cost'][index] = x
    except IndexError:
        pass
customer_demographic.head()
```
